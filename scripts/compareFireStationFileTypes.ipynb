```python
import os
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
# read the fire stations data in all 3 forms to compare
csv = gpd.read_file('Fire_Stations.csv')
geojson = gpd.read_file('Fire_Stations.geojson')
shp = gpd.read_file('Fire_Stations.zip')
```

```python
print('csv',csv.crs)
print('geojson',geojson.crs)
print('shapefile',shp.crs)
```

```python
# comparing shp and geojson:
difCols = [i for i in range(len(geojson.columns)) if not shp.iloc[:,i].equals(geojson.iloc[:,i])]
print(shp.columns[difCols])
print(geojson.columns[difCols])
difCols
```

```python
print(shp[shp.iloc[:,2] != geojson.iloc[:,2]].iloc[:,2])
print(geojson[shp.iloc[:,2] != geojson.iloc[:,2]].iloc[:,2])
```

```python
print(shp.columns[difCols])
print(shp.columns[shp.isna().sum() > 0])
print(geojson.columns[difCols])
print(geojson.columns[geojson.isna().sum() > 0])
```

```python
print(shp['geometry'].distance(geojson.to_crs(shp.crs)['geometry']).describe())
```
