```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import gudhi as gd
```

```python
# create a grid
xvals = np.arange(0,10,0.05)
yvals = np.arange(0,10,0.05)
nx = len(xvals)
ny = len(yvals)

grid = np.array([[x,y] for x in xvals for y in yvals])
```

```python
# f = lambda x,y: (x-5)**2 + (y-5)**2
f = lambda x,y: 50 - ((x-5)**2 + (y-5)**2)
# f = lambda x,y: np.cos(0.04*np.pi*((x-5)**2+(y-5)**2))
filt_vals = [f(x,y) for x,y in grid]
```

```python
plt.scatter(grid[:,0],grid[:,1],c=filt_vals)
plt.colorbar()
```

```python
cubicalComplex = gd.CubicalComplex(
    dimensions = [nx,ny],
    top_dimensional_cells = filt_vals
)
```

```python
cubicalComplex.dimension()
```

```python
cubicalComplex.num_simplices()
```

```python
barCodes = cubicalComplex.persistence()
```

```python
barCodes
```

```python
gd.plot_persistence_diagram(barCodes)
```
