```python
both = arc.index.intersection(chi.index)
len(both)
```

```python
# let's test and see if for a single geoid in both data sources
# that the relevant info (basically just the geometry) matches
idx = both[0]
print(arc.loc[idx])
print('chi')
print(chi.loc[idx])
```

```python
# Looking at this, it seems they match. Looking at columns, it seems like we can match columns
xwalk = {
    'STATE':'statefp10',
    'NAME':'name10',
    'BLOCK':'blockce10',
    'TRACT':'tractce10',
    'COUNTY':'countyfp10'
}
cols = list(xwalk.values())
arc.rename(columns=xwalk).loc[both,cols].equals(chi.loc[both,cols])
```

```python
# Okay! This is good. But do the geometry's of the census blocks contained in both sources match?
arc = arc.to_crs('epsg:26916') # I got this CRS from: https://gis.stackexchange.com/questions/346712/choosing-crs-for-chicago-illinois
chi = chi.to_crs('epsg:26916')
```

```python
dists = arc.loc[both].centroid.distance(chi.loc[both].centroid)
```
